# 2. Téléchargement des données

On récupère les données en ligne depuis le lien fourni qui contient les données sur lesquelles nous allons travailler pour l'analyse des modèles de recommandation et aussi pour notre MVP.<br>
Les données sont stockées dans un dossier `data` en local.

```python
# Folder where the data will be stored
DATA_DIR = ROOT_DIR / "data"
# Zip file URL to download
ZIP_URL = "https://s3-eu-west-1.amazonaws.com/static.oc-static.com/prod/courses/files/AI+Engineer/Project+9+-+R%C3%A9alisez+une+application+mobile+de+recommandation+de+contenu/news-portal-user-interactions-by-globocom.zip"


# Add functions to download and extract the zip file
@retry(
    stop=stop_after_attempt(5),
    wait=wait_exponential(multiplier=1, min=2, max=10),
    retry=retry_if_exception_type(requests.exceptions.RequestException),
    reraise=True,
)
def download_and_extract_zip(url, extract_to):
    """Download a zip file from a URL and extract it to a specified directory, with retry."""
    try:
        response = requests.get(url)
        response.raise_for_status()
        zip_path = os.path.join(extract_to, "data.zip")
        with open(zip_path, "wb") as f:
            f.write(response.content)
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(extract_to)
        os.remove(zip_path)
        if os.path.exists(os.path.join(extract_to, "clicks_sample.csv")):
            os.remove(os.path.join(extract_to, "clicks_sample.csv"))
    except requests.exceptions.RequestException as e:
        raise e
    except zipfile.BadZipFile as e:
        raise e


def unzip_clicks_data(extract_to):
    """Unzip clicks data from the specified directory."""
    zip_path = os.path.join(extract_to, "clicks.zip")
    if not os.path.exists(zip_path):
        return
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_to)
    os.remove(zip_path)
```

```python
# Download and extract the zip file in the data directory
if not DATA_DIR.exists():
    DATA_DIR.mkdir(parents=True, exist_ok=True)
download_and_extract_zip(ZIP_URL, DATA_DIR)
unzip_clicks_data(DATA_DIR)
```

## 3.1 clicks.zip


Je vais effectuer la reconstruction sous forme de dataframe de l'ensemble des données de clics, en les chargeant à partir des fichiers CSV contenus dans le fichier zip `clicks.zip`. Chaque fichier CSV représente les interactions des utilisateurs pour une heure donnée, et nous allons les combiner pour créer un seul dataframe contenant toutes les interactions.

```python
# List every file in the clicks directory with glob
clicks_files = glob(str(DATA_DIR / "clicks" / "*.csv"))
n_hours = len(clicks_files)
print(f"Number of hours of clicks data: {n_hours}")
print(f"Number of days of clicks data: {n_hours // 24}")

# Read all CSV files into a list of DataFrames
clicks_dfs = [pd.read_csv(file) for file in clicks_files]
# Concatenate all DataFrames into a single DataFrame
clicks_df = pd.concat(clicks_dfs, ignore_index=True)
print(f"Shape of concatenated clicks DataFrame: {clicks_df.shape}")
```

- **clé primaire** : identifiant unique (parent/child selon la hiérarchie)

- **catégorielle** : variable qualitative
- **timestamp** : date/heure

Le dataframe pèse plus de 273 Mo et les data types ne sont pas optimisés. Il est donc nécessaire de les optimiser pour réduire la taille du dataframe et améliorer les performances des opérations de traitement des données.

```python
# Define timestamps columns
timestamp_cols = ["session_start", "click_timestamp"]

# Convert timestamp columns to datetime and numeric types to best integer type to save memory
clicks_df = clicks_df.assign(
    **{
        col: lambda df, col=col: pd.to_datetime(df[col], unit="ms")
        for col in timestamp_cols
    }
).assign(
    **{
        col: lambda df, col=col: pd.to_numeric(df[col], downcast="integer")
        for col in clicks_df.columns
        if col not in timestamp_cols
    }
)
```

**Analyse de la variable `session_id`**: Unique identifier of the session (anonymized).

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 10), gridspec_kw={"width_ratios": [3, 1]})

session_counts = clicks_df.groupby("session_id").size()
session_counts_truncated = session_counts.pipe(lambda s: s[(s <= s.quantile(0.999))])
fig.suptitle(
    "Distribution of number of clicks per session (raw and truncated)",
    fontsize=16,
    fontweight="bold",
)
for i, df in enumerate([session_counts, session_counts_truncated]):
    df.value_counts().sort_index().plot(
        kind="bar",
        title=(
            "Number of clicks per session (all sessions)"
            if i == 0
            else "Number of clicks per session (truncated from 2 clicks per sessions min to quantile 0.999)"
        ),
        ax=ax[i, 0],
        color="skyblue",
    )
    ax[i, 0].set_xlabel("Session size (number of clicks)")
    ax[i, 0].set_ylabel("Count of sessions")
    ax[i, 0].grid(axis="y")
    stats_text = "\n".join(
        [f"{key}: {value:.1f}" for key, value in df.describe().items()]
    )
    stats_text = "Descriptive statistics:\n\n" + stats_text
    ax[i, 1].text(
        0.0,
        1.0,
        stats_text,
        transform=ax[i, 1].transAxes,
        fontsize=10,
        fontweight="bold",
        verticalalignment="top",
        horizontalalignment="left",
    )
    ax[i, 1].axis("off")
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

**Analyse de la distribution brute du nombre de clics par session :**
- une majorité de sessions très courtes avec une médiane à 2 clics sur l'ensembe des 1 048 594 sessions
- une distribution avec une longue queue à droite, révelant des outliers, et des valeurs extrêmes avec des sessions allant jusqu'à 124 clics (bot?)
- Tronquée au 0.999 quantile, on écarte les sessions avec plus de 20 clics, ce qui reste un nombre élevé pour une session humaine.
- Deux approches possibles pour la suite seraient :
    - soit d'écarter toutes sessions avec moins de 5 clics pour assurer d'avoir min 5 clics dans la session de test
    - soit de conserver toutes les sessions et après train-test split, de filtrer les sessions de test avec moins de 5 clics. Cette dernière approche permettrait de conserver les sessions avec moins de 5 clics dans l'ensemble d'entraînement, ce qui pourrait être utile pour la modélisation et améliorer la robustesse du modèle.

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6), gridspec_kw={"width_ratios": [3, 1]})

df = clicks_df["click_article_id"].value_counts(ascending=False).reset_index(drop=True)
cumsum_pct = df.cumsum() / df.sum() * 100

# Prepend 0 to x and y for the cumsum curve
x = np.insert(df.index.values, 0, 0)
y = np.insert(cumsum_pct.values, 0, 0)

fig.suptitle(
    "Distribution of Clicks per Article",
    fontsize=16,
    fontweight="bold",
)
ax[0].plot(
    x,
    y,
    drawstyle="steps-post",
    color="skyblue",
)
ax[0].set_xlabel("Number of Top Articles")
ax[0].set_ylabel("Cumulative Percentage of Clicks (%)")
ax[0].set_title("Cumulative Percentage of Clicks by Top Articles")
ax[0].set_ylim(0, 100)
ax[0].grid()

# Find the number of articles that account for 50%, 90%, 95% & 99% of clicks
for pct, color in zip(
    [0.5, 0.9, 0.95, 0.99],
    ["orange", "red", "purple", "blue"],
):
    n_articles = (cumsum_pct < pct * 100).sum() + 1
    threshold = df.iat[n_articles - 1]
    ax[0].axvline(
        n_articles,
        color=color,
        linestyle="--",
        label=f"{n_articles} articles = {int(pct * 100)}% of clicks (threshold = {threshold})",
    )
ax[0].legend()
# Set the y-axis to show percentage format
ax[0].yaxis.set_major_formatter(PercentFormatter())

# Display descriptive statistics for the clicks DataFrame
stats_text = "\n".join([f"{key}: {value:.1f}" for key, value in df.describe().items()])
stats_text = "Descriptive Statistics:\n\n" + stats_text

ax[1].text(
    0.0,
    1.0,
    stats_text,
    transform=ax[1].transAxes,
    fontsize=10,
    fontweight="bold",
    verticalalignment="top",
    horizontalalignment="left",
)
ax[1].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
clicks_df = clicks_df.merge(
    clicks_df["click_article_id"]
    .value_counts(normalize=True)
    .rename("article_popularity")
    .reset_index(),
    on="click_article_id",
    how="left",
).merge(
    clicks_df["click_article_id"]
    .value_counts(normalize=False)
    .rename("article_count")
    .reset_index(),
    on="click_article_id",
    how="left",
)


clicks_df = clicks_df.sort_values(
    ["session_start", "click_timestamp"], ascending=True
).assign(
    **{
        "read_duration": lambda df: df.groupby("session_id")["click_timestamp"]
        .diff()
        .shift(-1)
        .dt.total_seconds()
        .replace({30.0: np.nan}),
        "click_ranking": lambda df: df.groupby("session_id")["click_timestamp"]
        .rank(method="first")
        .astype(int),
    }
)

# Calcul de la durée moyenne par article (en excluant les valeurs manquantes)
median_duration_per_article = clicks_df.groupby("click_article_id")[
    "read_duration"
].median()

# Imputation de la durée manquante par la moyenne de l'article correspondant
clicks_df["read_duration"] = (
    clicks_df["read_duration"]
    .fillna(clicks_df["click_article_id"].map(median_duration_per_article))
    .fillna(median_duration_per_article.median())
)
```

```python
cum_pop = (
    clicks_df.groupby("click_article_id")["article_popularity"]
    .first()
    .sort_values(ascending=False)
    .cumsum()
    .reset_index(drop=True)
)

quantiles = np.arange(0.1, 1.01, 0.1)
quantile_indices = (quantiles * (len(cum_pop) - 1)).astype(int)
quantile_labels = [f"{q:.1f}" for q in quantiles]

quantiles_ = np.array([0.02, 0.04, 0.06, 0.08])
quantile_indices_ = (quantiles_ * (len(cum_pop) - 1)).astype(int)
quantile_labels_ = [f"{q:.2f}" for q in quantiles_]

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(
    cum_pop.index,
    cum_pop.values,
    color="deepskyblue",
    label="Cumulative article popularity",
)
ax.fill_between(
    cum_pop.index,
    cum_pop.values,
    color="lightblue",
    alpha=0.7,
)
for idx, quant in enumerate(quantiles_):
    ax.axvline(
        quantile_indices_[idx],
        color="red",
        linestyle="--",
        label=f"{quantile_labels_[idx]} quantile",
    )
ax.set_title("Cumulative distribution of article popularity")
ax.set_xlabel("Quantile")
ax.set_ylabel("Cumulative article popularity")
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_xticks(np.concatenate((quantile_indices_, quantile_indices)))
ax.set_xticklabels(quantile_labels_ + quantile_labels, rotation=45)
ax.grid(axis="both")
ax.legend()
plt.tight_layout()
plt.show()
```

**Observations :**
- Une minorité d'articles génère la majorité des clics, avec une longue queue à droite de la distribution.
- Cela peut constituer une bonne baseline pour un modèle de recommandation basé sur la popularité mais aussi de cold start pour les nouveaux utilisateurs, mais il faudra aussi prendre en compte les articles moins populaires pour éviter de biaiser les recommandations.

# 5. Filtrage des données pour l'entraînement et l'évaluation
On va écarter tous les outliers qui pourraient bruiter les résultats de l'entraînement et de l'évaluation des modèles.

```python
merged_df = pd.read_pickle(DATA_DIR / "merged_df.pickle")
print("Merged DataFrame shape:", merged_df.shape)

if not merged_df["click_timestamp"].is_monotonic_increasing:
    merged_df = merged_df.sort_values("click_timestamp").reset_index(drop=True)

print(
    "merged_df is sorted by click_timestamp (oldest to newest):",
    merged_df["click_timestamp"].is_monotonic_increasing,
)
print("Merged DataFrame shape:", merged_df.shape)
display(merged_df.head())
```

```python
assert merged_df.isnull().sum().sum() == 0
assert merged_df.duplicated().sum() == 0
```

**Voici l'ensemble des masques appliqués pour filtrer les données :**

```python
# Filtre 1 - Outliers : On écarte les clics après le 17 octobre 2017
mask_1 = merged_df["click_timestamp"] <= "2017-10-17 00:00:00"
# Filtre 2 - Outliers : On écarte les articles qui ont moins de 30 clics
articles_masked = merged_df["article_id"].value_counts()
valid_article_ids = articles_masked[articles_masked >= 30].index
mask_2 = merged_df["article_id"].isin(valid_article_ids)
# Filtre 3 - Outliers : On écarte les sessions de plus de 20 clics
mask_3 = merged_df["session_size"] <= 20
# Filtre 4 - Outliers : On écarte les utilisateurs qui ont cliqué moins de 2 fois sur le même article représentant aussi des outliers dans notre jeu de données
users_masked = (
    merged_df.groupby(["user_id", "click_article_id"])
    .size()
    .reset_index(name="click_count_same_article")
)
valid_users_ids = users_masked[users_masked["click_count_same_article"] >= 2][
    "user_id"
].unique()
mask_4 = merged_df["user_id"].isin(valid_users_ids)
# Filtre 5 - Outliers : On conserve  les utilisateurs avec un minimum de 6 clics (au moins 5 pour le test et 1 pour l'entraînement) et un maximum de 67 clics (pour éviter les utilisateurs extrêmes qui pourraient fausser les résultats)
users_masked = merged_df.groupby("user_id")["click_timestamp"].size()
mask_5 = merged_df["user_id"].isin(
    users_masked[(users_masked > 6) & (users_masked <= 67)].index
)
# Apply all filters
filtered_df = merged_df[mask_1 & mask_2 & mask_3 & mask_4 & mask_5].copy()
print("Filtered DataFrame shape:", filtered_df.shape)
```

**Voici la stratégie que l'on va suivre pour séparer les données :**
1. Choisir un range de dates par jour à tester pour le split et slider temporellement.
2. Calculer le nombre de clics par utilisateur dans le jeu de test et le jeu d'entraînement avec .groupby("user_id")["click_timestamp"].size()
3. Récupérer les index des utilisateurs qui ont au moins 5 clics dans le jeu de test et au moins 1 clic dans le jeu d'entraînement
4. Filtrer les utilisateurs dans le jeu de test et le jeu d'entraînement en fonction des index récupérés
5. Calculer combien il reste d'utilisateurs valides par split et aussi le nombre de clicks par utilisateur dans le jeu de test et le jeu d'entraînement pour comparer ensuite les distributions ou juste la moyenne / mediane / etc.

```python
candidate_dates = pd.date_range(
    start=filtered_df["click_timestamp"].dt.date.min(),
    end=filtered_df["click_timestamp"].dt.date.max(),
    freq="D",
)
```

```python
results_dict = defaultdict(list)
# Iterate over each candidate date to create train/test splits
for date in candidate_dates[1:]:
    # Filter the DataFrame for the current date
    train = filtered_df[filtered_df["click_timestamp"] < date]
    test = filtered_df[filtered_df["click_timestamp"] >= date]

    # Compute the number of clicks by user in the training and test sets
    train_clicks = train.groupby("user_id")["click_timestamp"].size()
    test_clicks = test.groupby("user_id")["click_timestamp"].size()

    # Identify users with at least 1 click in the training set and at least 5 clicks in the test set
    valid_users = train_clicks[train_clicks >= 1].index.intersection(
        test_clicks[test_clicks >= 5].index
    )

    # Filter the training and test sets to only include valid users
    train_filtered = train[train["user_id"].isin(valid_users)]
    test_filtered = test[test["user_id"].isin(valid_users)]

    # Check if train and test filtered have the same user_id unique values
    if not set(train_filtered["user_id"].unique()).issubset(
        set(test_filtered["user_id"].unique())
    ):
        print(
            f"Warning: Some users in the training set ({len(train_filtered['user_id'].unique())}) are not present in the test set ({len(test_filtered['user_id'].unique())}) for date {date}."
        )
        continue

    # Compute the number of clicks by user in the training and test sets
    train_clicks = train_filtered.groupby("user_id")["click_timestamp"].size()
    test_clicks = test_filtered.groupby("user_id")["click_timestamp"].size()

    # Compute the number of unique users in the training and test sets
    # Store the results
    results_dict["date"].append(date)
    results_dict["valid_users"].append(len(valid_users))
    # Compute the average number of clicks per user in the training and test sets
    results_dict["train_avg_clicks"].append(train_clicks.mean())
    results_dict["test_avg_clicks"].append(test_clicks.mean())
    # Add the median number of clicks per user in the training and test sets
    results_dict["train_median_clicks"].append(train_clicks.median())
    results_dict["test_median_clicks"].append(test_clicks.median())
```

```python
results_df = pd.DataFrame(results_dict).set_index("date")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot clicks per user stats
results_df.filter(like="click", axis=1).plot(ax=axes[0])
axes[0].set_title("Clicks per User Statistics")
axes[0].grid(visible=True, axis="both", which="both")

# Plot valid users stats
results_df.filter(like="user", axis=1).plot(ax=axes[1])
axes[1].set_title("Valid Users Statistics")
axes[1].grid(visible=True, axis="both", which="both")

plt.tight_layout()
plt.show()
```

**Observations :**
- Les deux dates les plus appropriées pour effectuer le split sont le 10-09 et 10-10 2017. On obtient un nombre d'utilisateurs valides suffisant pour l'entraînement et l'évaluation des modèles.
- On enregistre sous forme de pickle les DataFrames filtrés pour les utiliser dans les notebooks d'entraînement et d'évaluation des modèles.

```python
for date_split in [pd.to_datetime("2017-10-09"), pd.to_datetime("2017-10-10")]:
    print(f"Train/test split for date: {date_split}")
    # Filter the DataFrame for the current date
    train = filtered_df[filtered_df["click_timestamp"] < date_split]
    test = filtered_df[filtered_df["click_timestamp"] >= date_split]

    # Compute the number of clicks by user in the training and test sets
    train_clicks = train.groupby("user_id")["click_timestamp"].size()
    test_clicks = test.groupby("user_id")["click_timestamp"].size()

    # Identify users with at least 1 click in the training set and at least 5 clicks in the test set
    valid_users = train_clicks[train_clicks >= 1].index.intersection(
        test_clicks[test_clicks >= 5].index
    )

    # Filter the training and test sets to only include valid users
    train[train["user_id"].isin(valid_users)].to_pickle(
        DATA_DIR
        / f"train_filtered_{date_split.strftime('%Y-%m-%d').split('-')[-1]}.pickle"
    )
    test[test["user_id"].isin(valid_users)].to_pickle(
        DATA_DIR
        / f"test_filtered_{date_split.strftime('%Y-%m-%d').split('-')[-1]}.pickle"
    )
```
